Importing data and overview

```python
# phenotype data to get sample_type
import pandas as pd
phenotypes = pd.read_csv("TCGA.STAD.sampleMap_STAD_clinicalMatrix.tsv", sep='\t')
phenotypes.rename(columns={"sampleID": "sample"}, inplace=True)
phenotypes.head()
```

```python
# flip format of gene data to match with phenotype data. 
geneData = pd.read_csv("TCGA.STAD.sampleMap_HiSeqV2_PANCAN.tsv", sep='\t', low_memory=False, index_col=0, header=None)
geneData = geneData.transpose()
geneData.head()
```

```python
# combined data of samples with type AND gene data
df = pd.merge(geneData, phenotypes, on='sample', how='inner')
df.head(10)
```

Gene expression data heatmap for all of the samples
* can specify which genes you want to look at with startGene and numOfGenes (using .iloc; referring to index)

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# df1 -> sort data by sample
df1 = df.sort_values(by='sample').reset_index(drop=True)

# create plot 
fig, axes = plt.subplots(nrows=2, ncols=1, sharex=True, figsize=(15, 10), gridspec_kw={'height_ratios': [1, 2]})
plt.xticks(np.arange(0, 46, step=1))

# 1. graph of dist of sample types
sampleTypes = df1['sample_type'].value_counts(sort=False)
sampleTypes.to_frame().T.plot.barh(stacked=True, cmap='coolwarm', ax=axes[0])

# 2. gene expression heatmap
startGene = 1 # always start at 1, not 0
numOfGenes = 20 # max: 20531??

for i in range(numOfGenes): # to avoid TypeErrors and NA problems
    col = df1.columns[startGene+i]
    df1[col] = pd.to_numeric(df1[col], errors='coerce')
    df1[col] = df1[col].fillna(0)
    
all_gene_expression = df1.iloc[:, startGene:startGene+numOfGenes]
sns.heatmap(all_gene_expression.T, cmap='coolwarm', cbar_kws={'orientation': 'horizontal'}, ax=axes[1])

plt.show()
```

The top x most variable genes for EACH sample_type (tumor or normal tissue)

```python
def most_variable_genes(passed_df, passed_val):
    # note: this code will show random 9 samples ; Normal Tissue only has 9 so this will  only affect the tumor
    # doing this for a better comparison between normal/tumor
    
    passed_df.iloc[:, 1:] = passed_df.iloc[:, 1:].apply(pd.to_numeric, errors='coerce').fillna(0) 
        # exclude sample number & fill NA values
    
    passed_df.loc['variability'] = passed_df.iloc[:, 1:].apply(lambda x: np.nanvar(pd.to_numeric(x, errors='coerce')), axis=0) 
        # calculate variability; .var would not work :(
    passed_df.iloc[:, 1:] = passed_df.iloc[:, 1:].sort_values(by='variability', axis=1, ascending=False) 
        # sort values descending of variability values
    
    gene_expression = passed_df.iloc[:-1, 1:passed_val]
        # top passed_val most variable genes
    gene_expression = gene_expression.sample(n=9, axis=0) 
        # pick random 9
    gene_expression = gene_expression.apply(pd.to_numeric) 
        # fixing error; values were not being passed as numerical when pulling from passed_df
    return gene_expression
```

```python
# create plot
fig, axes = plt.subplots(nrows=1, ncols=2, sharex=False, figsize=(20, 8))
plt.xticks(np.arange(0, 10, step=1)) # change for num of samples to show

# df2 -> subset of data, tumor tissue; the most variable genes for tumor
df2 = df[df['sample_type'] == 'Primary Tumor'].iloc[:,:20531].copy()
# df3 -> subset of data, solid tissue normal; the most variable genes for normal tissue
df3 = df[df['sample_type'] == 'Solid Tissue Normal'].iloc[:,:20531].copy()

# top 15 most variable genes for each type
fig.suptitle('15 most variable genes for each sample type', fontsize=18)

# create plots
# 1. tumor
gene_expression_tumor = most_variable_genes(df2, 15)
# 2. normal
gene_expression_normal = most_variable_genes(df3, 15)

scale_min = min(gene_expression_tumor.min().min(), gene_expression_normal.min().min())
scale_max = max(gene_expression_tumor.max().max(), gene_expression_normal.max().max())

sns.heatmap(gene_expression_tumor.T, cmap='coolwarm', cbar_kws={'orientation': 'horizontal'}, vmin=scale_min, vmax=scale_max, ax=axes[0])
axes[0].set_title('Primary Tumor')

sns.heatmap(gene_expression_normal.T, cmap='coolwarm', cbar_kws={'orientation': 'horizontal'}, vmin=scale_min, vmax=scale_max, ax=axes[1])
axes[1].set_title('Solid Tissue Normal')

plt.show()
```
